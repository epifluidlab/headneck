```python
from graphviz import Digraph

dot = Digraph(comment='Sample Filtering Process')

# Define a refined node style with a cohesive color scheme
node_style = {'shape': 'rect', 'style': 'filled,rounded', 'fillcolor': '#AED6F1', 'fontname': 'Helvetica', 'fontsize': '12'}
end_node_style = {'shape': 'rect', 'style': 'filled,rounded', 'fillcolor': '#AED6F1', 'fontname': 'Helvetica', 'fontsize': '12'}

# Define nodes with descriptive text
dot.node('A', 'Total Samples at Collection\n185', **node_style)
dot.node('B', 'Samples Passing QC\n176', **node_style)
dot.node('C', 'Analysis Set\n151', **node_style)
dot.node('D', 'Pre-Holdout Set\n(Label Blind, Randomly Selected)\n25', **end_node_style)

# Add edges with labeled reductions
edge_style = {'fontsize': '10', 'fontname': 'Helvetica', 'color': '#2E86C1'}
dot.edge('A', 'B', label=' -9', **edge_style)
dot.edge('B', 'C', label='‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎  ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ‎ ', **edge_style)
dot.edge('B', 'D', label='', **edge_style)

# Render to PDF with cleanup disabled
output_path = 'SF1A'
dot.render(output_path, format='pdf', cleanup=False)

print(f'Graph rendered to {output_path}.pdf')
```
