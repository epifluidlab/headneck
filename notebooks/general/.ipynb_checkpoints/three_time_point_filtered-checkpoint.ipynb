```python
import pandas as pd
import matplotlib.pyplot as plt
from upsetplot import UpSet, from_memberships

csv_file_path = "/projects/b1198/epifluidlab/ravi/0130/headneck/metadata/RAW_HNSCC_METADATA.csv"
valid_ids_path = "/projects/b1198/epifluidlab/ravi/0130/headneck/notebooks/valid_ids.txt"

# Load the metadata CSV file
metadata_df = pd.read_csv(csv_file_path)

# Load valid IDs from the text file
with open(valid_ids_path, "r") as f:
    valid_ids = {line.strip() for line in f}  # Read lines and store as a set for fast lookup

# Filter metadata DataFrame to keep only valid IDs
metadata_df['ID'] = metadata_df['ID'].str.replace('-', '_', regex=False)
metadata_df = metadata_df[metadata_df["ID"].astype(str).isin(valid_ids)]


# Select specific columns
metadata_df = metadata_df[["ID", "Treatment Response", "Type of Visit", "Patient Number"]]


# Create a DataFrame that categorizes patients based on their visit types
patient_visits = pd.DataFrame({
    'Screen': metadata_df.groupby('Patient Number')['Type of Visit'].apply(lambda x: 'Screen' in x.values),
    'Day 0': metadata_df.groupby('Patient Number')['Type of Visit'].apply(lambda x: 'Day 0' in x.values),
    'Adj Wk 1': metadata_df.groupby('Patient Number')['Type of Visit'].apply(lambda x: 'Adj Wk 1' in x.values)
})

# Convert the DataFrame to membership format suitable for the UpSet plot
memberships = []
for idx, row in patient_visits.iterrows():
    sets = []
    if row['Screen']:
        sets.append('Screen')
    if row['Day 0']:
        sets.append('Day 0')
    if row['Adj Wk 1']:
        sets.append('Adj Wk 1')
    memberships.append(sets)

# Generate the data for the UpSet plot
upset_data = from_memberships(memberships)

# Create an UpSet plot
upset = UpSet(upset_data, subset_size='count', show_counts='%d', facecolor="lightblue")
upset.plot()

# Save the plot before showing it
plt.savefig('three_tp_dist_filtered.png', dpi=1000, bbox_inches='tight')

# Show the plot
plt.show()
```
