```python
bed_path = "../../Supplementary_Tables/ST3/rMDS_cluster3.tsv"
df = pd.read_csv(bed_path, sep='\t', header=None, names=['chrom', 'start', 'end'])
df.sort_values(by=['chrom', 'start'], inplace=True)

roi = []
chain_start = None
chain_count = 0
prev = None

for idx, row in df.iterrows():
    chrom, start, end = row

    if prev is not None:
        prev_chrom, prev_start, prev_end = prev
        if chrom == prev_chrom and start == prev_end:
            if chain_start is None:
                chain_start = [prev_chrom, prev_start]
                chain_count = 2  # includes prev and current
            else:
                chain_count += 1
        else:
            if chain_start:
                roi.append([prev_chrom, chain_start[1], prev_end, chain_count])
                chain_start = None
                chain_count = 0
    else:
        chain_count = 0
    prev = [chrom, start, end]

if chain_start:
    roi.append([prev[0], chain_start[1], prev[2], chain_count])

roi_df = pd.DataFrame(roi, columns=['chrom', 'start', 'end', 'merged_count'])
roi_df.sort_values(by='merged_count', ascending=False, inplace=True)

roi_df.to_csv("F2F.merged.rMDS.cluster3.bed", sep='\t', index=False, header=False)
```

```python
keyfunc = lambda x: tuple(int(n) for n in re.findall(r'\d+', x))
responder = hc_data.get_raw_features(hc_data.get_subset(hc_data.responder_ids))

sorted_cols = sorted(responder.columns, key=keyfunc)
_COL_RE = re.compile(r'chr([^:]+):(\d+)-(\d+)')

def select_interval_cols(chrom, start, end, overlap=False):
    """
    Given a list of sorted column names like
      ['chr10:1000000-1500000', 'chr10:10000000-10500000', ...],
    return the subset that lie in [start, end] on chromosome `chrom`.

    Parameters
    ----------
    sorted_cols : list of str
        The already-sorted list of "chr<id>:<start>-<end>" strings.
    chrom : str
        Chromosome identifier (e.g. "10").
    start : int
        Query interval start (inclusive).
    end : int
        Query interval end (inclusive).
    overlap : bool, default False
        If False, select columns fully contained in [start, end].
        If True, select any columns that overlap the interval at all.

    Returns
    -------
    list of str
        The columns from `sorted_cols` matching the criteria, in original order.
    """
    selected = []
    for col in sorted_cols:
        m = _COL_RE.match(col)
        if not m:
            continue
        col_chr, col_start, col_end = m.group(1), int(m.group(2)), int(m.group(3))
        if col_chr != chrom:
            continue
        if overlap:
            # any overlap
            if col_end >= start and col_start <= end:
                selected.append(col)
        else:
            # fully contained
            if col_start >= start and col_end <= end:
                selected.append(col)
    return selected
```

```python
import marsilea as ma
import marsilea.plotter as mp
tracks = df
palette = sns.color_palette("colorblind")
colors = {
    "Responder": palette[0],
    "Non-Responder": palette[5]
}

lims = {
    "Screen": 0.75,
    "Day 0": 0.75,
    "Adjusted Week 1": 0.75,
}

TRACK_HEIGHT = 0.5
TRACK_PAD = 0.3
canvas = ma.ZeroHeight(10, name="track")

for _, row in tracks.iterrows():
    cond, enz, track = row
    name = f"{cond}{enz}"
    color = colors[cond]
    canvas.add_bottom(
        mp.Area(
            track,
            color=color,
            add_outline=False,
            alpha=1,
            label=cond,
            label_loc="right",
        ),
        size=TRACK_HEIGHT,
        pad=TRACK_PAD,
        name=name,
    )


enz_canvas = ma.ZeroHeight(1, name="enz")
for enz in tracks["enz"].drop_duplicates():
    enz_canvas.add_bottom(
        mp.Title(f"  {enz}", align="left"),
        size=TRACK_HEIGHT * 2,
        pad=TRACK_PAD * 2,
        name=enz,
    )

comp = canvas + 1.5 + enz_canvas
comp.render()

for enz in tracks["enz"].drop_duplicates():
    ax = comp.get_ax("enz", enz)
    ax.axvline(x=0, color="k", lw=4)

# Modify the limits
for _, row in tracks.iterrows():
    name = f"{row['cond']}{row['enz']}"
    lim = lims[row["enz"]]
    ax = comp.get_ax("track", name)
    ax.set_ylim(-1*lim, lim)
    ax.set_yticks([-1*lim,lim])
    
comp.save('F2G_2.pdf')
```

```python
import numpy as np

def get_end_motifs(intervals, id_name):
    file_path = f"/projects/b1198/epifluidlab/ravi/0401/headneck/data/hg38_frag_filtered/{id_name}.hg38.frag.interval_end_motifs.500kb.tsv"
    intervals_set = set(intervals)
    results = []
    results_final = []
    with open(file_path) as f:
        header = f.readline().rstrip("\n").split("\t")
        for line in f:
            parts = line.rstrip("\n").split("\t")
            key = (parts[0], parts[1], parts[2])
            if key in intervals_set:
                results.append(line)
    for element in results:
        element = element.split("\t")
        count = int(element[4])
        arr = np.array(element[5:], dtype=np.float16) * count
        arr = np.rint(arr).astype(int)
        results_final.append(arr)
    results_final = np.array(results_final).sum(axis=0) / np.array(results_final).sum()
    return results_final

def get_end_motifs_set(intervals, id_names):
    results = []
    for element in id_names:
        results.append(get_end_motifs(intervals, element))
    return np.array(results).mean(axis=0)
```

```python
responder = get_end_motifs_set(intervals, responder)
non_responder = get_end_motifs_set(intervals, non_responder)

responder_screen = get_end_motifs_set(intervals, responder_screen)
non_responder_screen = get_end_motifs_set(intervals, non_responder_screen)
responder_day0 = get_end_motifs_set(intervals, responder_day0)
non_responder_day0 = get_end_motifs_set(intervals, non_responder_day0)
responder_adj_wk1 = get_end_motifs_set(intervals, responder_adj_wk1)
non_responder_adj_wk1 = get_end_motifs_set(intervals, non_responder_adj_wk1)
```

```python
from scipy.stats import f_oneway, mannwhitneyu
import scipy.stats as stats
from statsmodels.stats.multitest import multipletests

def create_volcano_plot(responder, nonresponder, ax, title, threshold=0.05):
    responder = (responder - responder.mean(axis=1, keepdims=True)) / responder.std(axis=1, keepdims=True)
    nonresponder = (nonresponder - nonresponder.mean(axis=1, keepdims=True)) / nonresponder.std(axis=1, keepdims=True)
    
    fold_changes = np.mean(nonresponder, axis=0) - np.mean(responder, axis=0)
    responder = responder
    nonresponder = nonresponder
    
    # Perform ANOVA and calculate p-values
    og_p_values = np.array([
        mannwhitneyu(responder[:, i], nonresponder[:, i], alternative='two-sided').pvalue
        for i in range(responder.shape[1])
    ])
    
    # Apply FDR correction
    _, corrected_p_values, _, _ = multipletests(og_p_values, alpha=threshold, method='fdr_bh')
    

    volcano_df = pd.DataFrame({
        'FoldChange': fold_changes,
        'PValue': og_p_values,
        'CorrectedPValue': corrected_p_values
    })
    
    volcano_df['NegLog10PValue'] = -np.log10(volcano_df['PValue'])
    threshold_log = -np.log10(threshold)
    
    volcano_df['Color'] = 'gainsboro' 
    volcano_df['Color'] = volcano_df.apply(
        lambda row: (
            'tomato' if row['FoldChange'] > 0.4 
            else 'cornflowerblue' if row['FoldChange'] < -0.4
            else 'gainsboro'
        ) if row['NegLog10PValue'] > threshold_log 
        else 'gainsboro',
        axis=1
    )
    volcano_df['PointSize'] = volcano_df['Color'].apply(lambda x: 10 if x != 'gainsboro' else 2)
    
    sns.scatterplot(
        data=volcano_df,
        x='FoldChange',
        y='NegLog10PValue',
        hue='Color',
        size='PointSize',
        sizes=(2, 10),
        palette={
            'cornflowerblue': 'cornflowerblue',
            'tomato': 'tomato',
            'gainsboro': 'gainsboro'
        },
        edgecolor=None,
        ax=ax,
        legend=False  # optional to keep it clean
    )
    
    ax.axhline(threshold_log, color='red', linestyle=(0, (3, 1, 1, 1, 1, 1)), linewidth=0.5)
    ax.axvline(x=0, color='gray', linestyle='--', linewidth=0.5)
    ax.axvline(x=-0.4, color='gray', linestyle='--', linewidth=0.2)
    ax.axvline(x=0.4, color='gray', linestyle='--', linewidth=0.2)
    
    ax.set_xlabel('End Motif Frequency\nZ-Score Difference\n(Non-Responder - Responder)', fontsize=8)
    ax.set_ylabel('-log10(FDR)', fontsize=8)
    ax.set_title(title, fontsize=8)
    ax.set_xlim(-1, 1)
    ax.set_ylim(0,6)
    ax.legend().remove()
    
    significant_indices = []
    for idx, row in volcano_df.iterrows():
        if (row['NegLog10PValue'] > threshold_log) and (abs(row['FoldChange']) > 0.4):
            index_label = idx.split(":")[0]
            significant_indices.append(idx)
            
            # Add text annotation
            if index_label!='CCAG':
                ax.text(
                    row['FoldChange']+0.1, row['NegLog10PValue'], index_label,
                    fontsize=6, ha='center', va='bottom', color='black', rotation=0
                )
            else:
                ax.text(
                    row['FoldChange']-0.1, row['NegLog10PValue'], index_label,
                    fontsize=6, ha='center', va='bottom', color='black', rotation=0
                )
                
    return significant_indices

# Plotting
fig, axes = plt.subplots(1, 1, figsize=(2, 3.5), dpi=300, sharey=True)
significant_indices = create_volcano_plot(responder, non_responder, axes, 'All Time Points')
axes.spines['right'].set_visible(False)
axes.spines['top'].set_visible(False)
axes.tick_params(axis='x', labelsize=8)
axes.tick_params(axis='y', labelsize=8)
plt.tight_layout()
plt.savefig('F2G.P1.v1.pdf', dpi=1000)
```
